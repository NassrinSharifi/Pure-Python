### Modified K

```python
modified_k = np.array([])
modified_tag = pt.produce_tag(16, ["x", "y"])
# print(total_tag_matrix)
# print(modified_tag)
# print(len(modified_tag))
for k, v in enumerate(total_tag_matrix):
    if v in modified_tag:
        modified_k = np.append(modified_k, K[k])

print(modified_k.reshape((32, 32)))
```

### Displacements

```python
modified_k = modified_k.reshape((32, 32))
modified_k_inv = np.linalg.inv(modified_k)
F = np.zeros(32)
F[24] = 1500
F[4] = 1500
F[5] = -1000
F[7] = -1000
F[9] = -1000
F[11] = -1000
F = F.reshape((32, 1))
displacement = np.dot(modified_k_inv, F)
print(displacement)
```
